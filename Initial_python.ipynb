```python
import pandas as pd
```

```python
data = pd.read_csv("Electric_Vehicle_Population_Data.csv")
```

```python
data = data.dropna()
data
```

```python
years = [2018,2019,2020,2021,2022,2023]
cutyears=data.loc[data["Model Year"] >= 2018]
cutyears
```

```python
zeroes=cutyears.loc[cutyears["Base MSRP"] ==0]
nonzero = cutyears.loc[cutyears["Base MSRP"] !=0]
```

```python
top10=list(cutyears.value_counts("Make").index[0:10])
top_df=cutyears.loc[cutyears["Make"].isin(top10)]
top_df
```

```python
top_df['Vehicle Location']=top_df['Vehicle Location'].apply(lambda x: str(x).replace("(","").replace(")","").replace("POINT ",""))
```

```python
top_df[['Longitude','Latitude']]=top_df['Vehicle Location'].str.split(' ', expand = True)
```

```python
top_df[['Longitude','Latitude']]=top_df[['Longitude','Latitude']].astype("float")
```
