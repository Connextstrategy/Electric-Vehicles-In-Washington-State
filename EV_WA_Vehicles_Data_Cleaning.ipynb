```python
import pandas as pd
```

```python
# Read Electric Vehicle Population file in from "https://data.wa.gov/Transportation/Electric-Vehicle-Population-Data/f6w7-q2d2"
data = pd.read_csv("Electric_Vehicle_Population_Data.csv")
```

```python
# Drop NAs from data for clean up 
data = data.dropna()
data
```

```python
# Drop all years less than 2022 as our model is based on new vehicle pricing 
cutyears=data.loc[data["Model Year"] >= 2022]
cutyears
```

```python
top10=list(cutyears.value_counts("Make").index[0:9])
#top_df=cutyears.loc[cutyears["Make"].isin(top10)]
top_df=cutyears
top_df
```

```python
# Used replacement syntax to change Vehicle Location verbage

top_df['Vehicle Location']=top_df['Vehicle Location'].apply(lambda x: str(x).replace("(","").replace(")","").replace("POINT ",""))
```

```python
# Split changed Vehicle in two different variables and changed them to 'float type'
top_df[['Longitude','Latitude']]=top_df['Vehicle Location'].str.split(' ', expand = True).astype("float")
top_df.head()
```

```python
# Dropped columns not required for analysis 
drop_df = top_df.drop(columns = ["Vehicle Location","2020 Census Tract","Electric Utility","DOL Vehicle ID","Legislative District",
                                "Electric Range","Base MSRP","State"])
drop_df.head()
```

```python
# Secured only models necessary for analysis after verifying Models on other database

drop_df=drop_df.loc[drop_df["Model"].isin(["C40",'EV6','I4','IONIQ 5','IONIQ 6','IX','KONA ELECTRIC','MODEL 3','MODEL S','MODEL X',
                                  'MODEL Y','NIRO','NIRO','XC40','MUSTANG MACH-E (AWD)'])]
```

```python
# Renamed items for ease of analysis and database creation
drop_df=drop_df.rename(columns = {"VIN (1-10)":"VIN","Postal Code":"Postal_Code","Model Year":"Model_Year",
                                 "Electric Vehicle Type":"Electric_Vehicle_Type",
                                  "Clean Alternative Fuel Vehicle (CAFV) Eligibility":"CAFV_Eligibility"})
drop_df.head()
```

```python
drop_df=drop_df.replace(
    {"C40": "Volvo C40 Recharge Twin Motor", 
     "EV6": "Kia EV6 GT", "I4": "BMW i4 eDrive35", "IONIQ 5": "Hyundai IONIQ 5 Long Range AWD",
     "IONIQ 6": "Hyundai IONIQ 6 Long Range AWD", "IX": "BMW iX M60", "KONA ELECTRIC": "Hyundai Kona Electric 65 kWh",
     "MODEL 3": "Tesla Model 3", "MODEL S": "Tesla Model S Plaid", "MODEL X": "Tesla Model X Dual Motor",
     "MODEL Y" : "Tesla Model Y", "NIRO" : "Kia Niro EV", "XC40" : "Volvo XC40 Recharge Twin Motor"})
drop_df
```

```python
# Renamed items for ease of analysis and database creation
drop_df.to_csv("EV_WA_cleaned_csv.csv",index=True)
```
